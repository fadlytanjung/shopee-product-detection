### Build Clasifier with Image AI

```python
from imageai.Prediction.Custom import ModelTraining

model_trainer = ModelTraining()
```

```python
model_trainer.setModelTypeAsDenseNet()
```

```python
model_trainer.setDataDirectory("./data")
```

```python
model_trainer.trainModel(num_objects=3, num_experiments=100, enhance_data=True, batch_size=32, show_network_summary=True)
```
